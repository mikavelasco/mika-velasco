#### "INVENTORY MAKER" CELL:

```python
# Check # comments before executing

################################################

inventory = {}

# ^ change inventory to the name of your inventory

command = input("Enter (Y) to add item to inventory:   ").upper()

while command == "Y":
    product_code = str(input("Enter product code:  "))
    product_name = str(input("Enter product name:  "))
    product_price = float(input("Enter product price:  "))
    product_quantity = int(input("Enter product quantity:  "))
    
    def enter_product(code, name, price, quantity):
        details = {}
        details["name"] = name
        details["price"] = price
        details["quantity"] = quantity
        entry = {code:details}
        inventory.update(entry)
          # ^ change inventory to the name of your inventory

        
    enter_product(product_code, product_name, product_price, product_quantity)
    command = input("Add another product? Y - Yes and N - End inventory edit:    ")

print("------------------------------------------")
print("Here is your inventory:")
print("Name of inventory: "+"inventory"+"\n")
                             # ^ change inventory to the name of your inventory in between the quotation marks
import pandas as pd

inventorydf = pd.DataFrame.from_dict(inventory, orient='columns')
                                    # ^ change inventory to the name of your inventory
# ^ change inventory to the name of your inventory

print(inventorydf)
        # ^ change inventory to the name of your inventory
```

### The Discount Randomizer

Want to give your customers a little bonus treat for spending more than ____ pesos in one purchase from your shop?

Use this discount randomizer program for your customers to use!

Note: run all the cells to test!

```python
import numpy as np
import random
import string
discount_code_list = []
```

```python
print("___________________________________________________________________________________________")
print("                                     Discount Giveaway                                     ")
print("Congrats! You have been given a chance to receive a 15% off discount on your next purchase.")
print("                    If you get the magic number of 5, you win the discount!                ")
print("___________________________________________________________________________________________")

start_randomizer = input("Type GO to start randomizing!  ")

magic_number = np.random.randint(1,10)

print("Your magic number is: ",magic_number)

def get_discount_code(length):
    letters = string.ascii_letters
    result = ''.join(random.choice(letters) for i in range(length))
    return result

discount_code = get_discount_code(8)

if magic_number == 5:
    print("Congratulations! Here is your discount code: ", discount_code)
    discount_code_list.append(discount_code)
    print("Please take note of your discount code so you can make use of it on your next purchase!")
else:
    print("Try again on your next purchase!")
```
